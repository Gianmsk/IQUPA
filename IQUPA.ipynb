```python
import os

# Directorios del dataset
image_dir = "dataset/train/images"  # Carpeta donde están las imágenes
label_dir = "dataset/train/labels"  # Carpeta donde están los archivos de etiquetas

# Clase a eliminar (por ejemplo, clase 7)
class_to_remove = 7

# Recorre todos los archivos de etiquetas
for label_file in os.listdir(label_dir):
    label_path = os.path.join(label_dir, label_file)
    image_path = os.path.join(image_dir, label_file.replace(".txt", ".jpg"))  # Ajusta la extensión si no es .jpg
    
    with open(label_path, "r") as file:
        lines = file.readlines()

    # Filtra las líneas que no tienen la clase a eliminar
    filtered_lines = [line for line in lines if int(line.split()[0]) != class_to_remove]

    if filtered_lines:
        # Si el archivo tiene líneas restantes, las reescribe
        with open(label_path, "w") as file:
            file.writelines(filtered_lines)
    else:
        # Si no quedan líneas, elimina el archivo de etiqueta y la imagen correspondiente
        os.remove(label_path)
        print(f"Etiqueta eliminada: {label_path}")
        if os.path.exists(image_path):
            os.remove(image_path)
            print(f"Imagen eliminada: {image_path}")

print("Proceso completado.")
```

```python
def preprocess_image(image_path):
    # Cargar imagen usando OpenCV
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convertir BGR a RGB
    image = cv2.resize(image, (192, 192))  # Redimensionar a 192x192
    image = image.astype(np.float32)  # Asegurarse de que la imagen esté en formato float32
    image = image / 255.0  # Normalizar la imagen
    image = np.expand_dims(image, axis=0)  # Añadir dimensión de batch
    return image

# Preparar la imagen
image_input = preprocess_image("xd1.jpg")

# Obtener detalles de los tensores
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

# Asignar el tensor de entrada
interpreter.set_tensor(input_details[0]['index'], image_input)

# Realizar la inferencia
interpreter.invoke()

# Obtener el resultado
output_data = interpreter.get_tensor(output_details[0]['index'])

# Mostrar el resultado
print("Resultados de la inferencia:", output_data)
```

```python
import matplotlib.pyplot as plt

predicted_class = np.argmax(output_data)

# Visualizar la imagen de entrada
plt.imshow(image_input[0])
plt.title(f"Predicción: Clase {predicted_class}")
plt.show()
```

```python
import torch
import torch.nn.utils.prune as prune
from ultralytics import YOLO

# Cargar el modelo de YOLO
model = YOLO("C:/Users/gianf/runs/detect/train36/weights/best.pt").model  # Cargar solo el modelo PyTorch

# Función para aplicar prunning a las capas convolucionales
def aplicar_prunning(modelo, cantidad_prunning):
    for nombre, módulo in modelo.named_modules():
        if isinstance(módulo, torch.nn.Conv2d):  # Prunning solo en capas convolucionales
            # Aplicar pruning L1 a los pesos
            prune.l1_unstructured(módulo, name='weight', amount=cantidad_prunning)

# Aplicar prunning con un porcentaje específico (por ejemplo, 50%)
aplicar_prunning(model, cantidad_prunning=0.7)

# Eliminar máscaras para mantener solo los pesos podados
for nombre, módulo in model.named_modules():
    if isinstance(módulo, torch.nn.Conv2d) and hasattr(módulo, 'weight_mask'):
        prune.remove(módulo, 'weight')

# Guardar el modelo podado
torch.save(model.state_dict(), 'model_p.pt')
```

```python
import os
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# Configuración
modelo_ruta = "C:/Users/gianf/runs/detect/train23/weights/best.pt"  # Reemplaza con la ruta de tu modelo entrenado
carpeta_imagenes = "C:/Users/gianf/yolo/test/images"  # Carpeta con tus imágenes
carpeta_resultados = "C:/Users/gianf/yolo/test/resuts"  # Carpeta donde se guardarán las imágenes con predicciones

# Crear carpeta de resultados si no existe
os.makedirs(carpeta_resultados, exist_ok=True)

# Cargar el modelo
model = YOLO(modelo_ruta)

# Procesar cada imagen en la carpeta
for imagen_nombre in os.listdir(carpeta_imagenes):
    imagen_ruta = os.path.join(carpeta_imagenes, imagen_nombre)
    imagen = cv2.imread(imagen_ruta)

    # Verificar que la imagen sea cargada correctamente
    if imagen is None:
        print(f"No se pudo cargar la imagen {imagen_nombre}.")
        continue

    # Realizar predicción
    results = model(imagen)

    # Dibujar cajas delimitadoras y etiquetas en la imagen
    for result in results[0].boxes:
        x1, y1, x2, y2 = map(int, result.xyxy[0])  # Coordenadas de la caja
        conf = result.conf.item()  # Convertir confianza a escalar
        clase_id = int(result.cls)  # ID de la clase

        # Obtener la etiqueta de clase usando el ID
        label = model.names[clase_id]  # Obtenemos el nombre de la clase desde 'model.names'

        # Dibujar caja y etiqueta
        cv2.rectangle(imagen, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(imagen, f"{label} {conf:.2f}", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Guardar la imagen con predicciones en la carpeta de resultados
    imagen_resultado_ruta = os.path.join(carpeta_resultados, imagen_nombre)
    cv2.imwrite(imagen_resultado_ruta, imagen)
```

```python
import tensorflow as tf

# Cargar el modelo TFLite
def load_tflite_model(tflite_file_path):
    with open(tflite_file_path, "rb") as f:
        tflite_model = f.read()
    return tflite_model

# Obtener información del modelo
def inspect_tflite_model(tflite_file_path):
    # Cargar el modelo TFLite
    interpreter = tf.lite.Interpreter(model_path=tflite_file_path)
    interpreter.allocate_tensors()  # Asignar los tensores

    # Obtener detalles de las entradas
    input_details = interpreter.get_input_details()
    print("Entradas:")
    for idx, input_tensor in enumerate(input_details):
        print(f" Input {idx}:")
        print(f"  - Nombre: {input_tensor['name']}")
        print(f"  - Forma: {input_tensor['shape']}")
        print(f"  - Tipo: {input_tensor['dtype']}")

    # Obtener detalles de las salidas
    output_details = interpreter.get_output_details()
    print("\nSalidas:")
    for idx, output_tensor in enumerate(output_details):
        print(f" Output {idx}:")
        print(f"  - Nombre: {output_tensor['name']}")
        print(f"  - Forma: {output_tensor['shape']}")
        print(f"  - Tipo: {output_tensor['dtype']}")

    # Inspeccionar los tensores intermedios (opcional)
    print("\nTensores intermedios:")
    tensor_details = interpreter.get_tensor_details()
    for tensor in tensor_details:
        print(f"  - Nombre: {tensor['name']}")
        print(f"  - Índice: {tensor['index']}")
        print(f"  - Forma: {tensor['shape']}")
        print(f"  - Tipo: {tensor['dtype']}")


# Ruta al archivo TFLite
tflite_file_path = "trained.tflite"

# Inspeccionar el modelo
inspect_tflite_model(tflite_file_path)
```
